fruits.sort(key=str.lower): This is the core of the sorting. The sort() method sorts the list elements in ascending order by default. However, the key=str.lower argument modifies this behavior. It tells Python to convert each string to lowercase temporarily for comparison purposes only before sorting. The actual items in the list remain with their original casing. This ensures that 'Apple' and 'apple' are treated the same for sorting, leading to a case-insensitive sort.

```python
fruits = ["Apple","Banana","Cherry","Mango","elderberry","Banana"]
fruits.sort(key=str.lower)
print(fruits)
```

```python
numbers = [1,2,3,4,5,1,2,3,123]
#slicing
print(numbers[2:6])
print(numbers[:5])
print(numbers[5:])
print(numbers[:])

#slicing with half
half1 = numbers[:len(numbers)//2]
half2 = numbers[len(numbers)//2:]
print(half1)
print(half2)
```
